```python
import math

D=32
# Input pipes
for i in range(D):
    print("hls::stream<item> pipe_in"+str(i)+";")
print("\n")
# Output pipes
for i in range(D):
    print("hls::stream<item> pipe_out"+str(i)+";")
```

```python
# interconnection input lx connections (disctance 2^x)
for x in range(int(math.log2(D))): #0,1,2,3,4 when D=32
    for i in range(D):
        if (0<=i%(2**(x+1))<2**x):
            print("hls::stream<item> pipe_l"+str(x)+"_"+str(i)+"to"+str(i)+";")
            print("hls::stream<item> pipe_l"+str(x)+"_"+str(i)+"to"+str(i+2**x)+";")
        if (2**x<=i%(2**(x+1))<2**(x+1)):
            print("hls::stream<item> pipe_l"+str(x)+"_"+str(i)+"to"+str(i-2**x)+";")
            print("hls::stream<item> pipe_l"+str(x)+"_"+str(i)+"to"+str(i)+";")  
    print("\n")
```

### Assumption: Y=2*D, Y>16

```python
fstr=""
for i in range(int(num_treebankgrounps)):
    fstr+="\t#pragma HLS aggregate variable=treemcts"+str(i)+"\n"

fstr+="\tprintf(\"In module out...\\n\");\n\tloop_pipeline_switch_out:while (1){\n\t\t#pragma HLS pipeline II=1\n"

p_in_strs=["up","down"]

target_bank_divider=int(Y/D)
for i in range(D):
    for instr in p_in_strs:
        fstr+="\t\tif (!inpipe"+str(i)+instr+".empty()){\n\t\t\titem new_sel; Tnode to_select;\n"
        fstr+="\t\t\titem from"+str(i)+instr+" = inpipe"+str(i)+instr+".read();\n"
        for j in range(target_bank_divider):
            global_bank_index=int(j*D+i)
            bank_group_number=int(global_bank_index/16)
            index_in_bank_group=int(global_bank_index%16)
            if (j==0):
                fstr+="\t\t\tif ((int)from"+str(i)+instr+".bankidx/"+str(target_bank_divider)+"=="+str(j)+") {to_select = treemcts"+\
                str(bank_group_number)+"[from"+str(i)+instr+".nodeidx].banks["+str(index_in_bank_group)+"];} //global_bank_index="+str(global_bank_index)+" \n"
            else:
                fstr+="\t\t\telse if ((int)from"+str(i)+instr+".bankidx/"+str(target_bank_divider)+"=="+str(j)+") {to_select = treemcts"+\
                str(bank_group_number)+"[from"+str(i)+instr+".nodeidx].banks["+str(index_in_bank_group)+"];} //global_bank_index="+str(global_bank_index)+" \n"
        fstr+="\t\t\tint new_bkidx; int new_neidx; fotype ndidx;\
        \n\t\t\tchildargmax(new_bkidx,new_neidx,ndidx,to_select); //assigned values to new_bkidx amd new_neidx\
        \n\t\t\tnew_sel.bankidx = new_bkidx; new_sel.nodeidx = new_neidx;\
        \n\t\t\tnew_sel.pipstage = from"+str(i)+instr+".pipstage+1; //e.g. pass the bank index output by stage 1 to be input by stage 2 in next iteration\
        \n\t\t\toutpipe0.write(new_sel);\n\t\t}\n"
    fstr+="\n"

print("\n")
print(fstr)
```

## loop_renew after switch_out
